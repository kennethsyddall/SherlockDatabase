```python
import selenium as sl
import requests
import os
import re
from selenium.webdriver.common.by import By
from selenium import webdriver
from bs4 import BeautifulSoup
from time import sleep
import urllib

lat_ctrs=['Antigua and Barbuda','Argentina','Bahamas','Barbados','Belize','Bolivia (Plurinational State of)','Brazil','Canada','Chile','Colombia',
    'Costa Rica','Cuba','Dominican Republic','Ecuador','El Salvador','United States of America','Grenada',
    'Guatemala','Guyana','Haiti','Honduras','Jamaica','Mexico','Nicaragua','Panama','Paraguay','Dominica','Peru',
    'Saint Kitts and Nevis','Saint Vincent and the Grenadines','Saint Lucia','Suriname','Trinidad and Tobago',
    'Uruguay','Venezuela (Bolivarian Republic of)']


##########################################################################################

def search_ctrs(url):
    driver = webdriver.Firefox()
    driver.get(url)
    sleep(15)
    if '/treaties/' in url:
        elem=driver.find_element_by_xpath('//a[@href="#country-treaty"]')
        elem.click()
    sleep(5)
    soup = BeautifulSoup(driver.page_source, 'lxml')
    non_ctrs = soup.find_all('a',class_="cover-parent")
    driver.quit()
    ctrs=[]
    for c in non_ctrs:
        ctr =c.get("href").split('"value":"')[1].split('"}]')[0]
        ctrs.append(ctr)
    return ctrs

##########################################################################################

def extract_urls(ctr,op):
    
    if op==1:
        op = "legdb"
        add = "legislation@country_label_s"
    elif op==2:
        op = "strategies"
        add = "strategy.country..country.name.html_s"
    elif op==3:
        op = "treaties"
        add = "countryTreatyStatus.countriesInvolved.country..country.name.html_s1"
    
    driver = webdriver.Firefox()
    
    url = 'https://sherloc.unodc.org/cld/v3/sherloc/'+op+'/search.html?lng=en#?c=%7B%22filters%22:%5B%7B%22fieldName%22:%22en%23'+add+'%22,%22value%22:%22'+ctr+'%22%7D%5D,%22sortings%22:%22%22%7D'
    
    con = 0
    while con == 0:
        try:
            driver.get(url)
            con = 1
        except:
            con = 0
            sleep(1)
    
    urls=[]
    found=10000
    while found != len(urls):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        soup = BeautifulSoup(driver.page_source, 'lxml')
    
        if op != 'legdb':
            nothf = soup.find('div',class_="ng-hide",id="nothing-found")
            if nothf == None:
                driver.quit()
                return []
        try:
            found = soup.find('div', id="found-count")
            found = int(found.find(text=True).split(' ')[1])
        except:
            continue
        tags = soup.find_all('a',class_="cover-parent")

        urls=[]
        for t in tags:
            url='https://sherloc.unodc.org'+t.get("href")
            if 'v3' not in url: 
                urls.append(url)
                
        
        
    driver.quit()
    return urls

##########################################################################################

def remplace_title_text(title,text):
    ctr_text={'«':'$<<$','»':'$>>$',"_":"\_","%":"\%","&":"\&",
              "$":"\\\$ ","":"•","":"•","":"•",'"':"'",'ñ':'ñ',
              'ó':'ó','á':'á','é':'é','í':'í','ú':'ú','Á':'Á',
              'É':'É','Í':'Í','Ó':'Ó','Ú':'Ú','−':'-','─':'-',
              '«':'$\ll$ ','»':'$\gg$ ','ê':'ê','ô':'ô','ã':'ã',
              'õ':'õ','à':'à','è':'è','ô':'ô','ç':'ç','â':'â',
              'ü':'ü','#':'\#','':'.',' ':'•',' ':'•','':'•',
              ';':';',' ':'•','`(':'(','а':'a','ï':'ï','û':'û',
              '}-':')-','{':'(','}':')','`S':'S','`E':'E',
              '':'$\rightarrow$ ','●':'•','\\':'',"`,":"´,"}
    
    
    ctr_text = dict((re.escape(k), v) for k, v in ctr_text.items())
    pattern_text = re.compile("|".join(ctr_text.keys()))
    
    ctr_title={"/":"_",'ñ':'ñ','ó':'ó','á':'á','é':'é',
               'í':'í','ú':'ú','Á':'Á','Í':'Í','Ó':'Ó','"':'',':':' '}
    ctr_title = dict((re.escape(n), m) for n, m in ctr_title.items())
    pattern_title = re.compile("|".join(ctr_title.keys()))
    
    new_text=[]
    for n,line in enumerate(text):
        if text[n]=='\n' or text[n]=='\xa0':
            new_text.append('')
            new_text.append('')
        text[n] = pattern_text.sub(lambda m: ctr_text[re.escape(m.group(0))], text[n])
        new_text.append(text[n])
    
    title = pattern_title.sub(lambda m: ctr_title[re.escape(m.group(0))], title)
    
    return title,new_text

##########################################################################################

def load_text(url):
    con=0
    while con==0:
        try:
            res = requests.get(url)
            con=1
        except:
            con=0
            sleep(1)
    html_page = res.content
    soup = BeautifulSoup(html_page, 'lxml')
    attach = []
    if '/legislation/' in url:
        title,text= ('epa',['salepa','quepacho'])
        cuad = soup.find('div',class_="case-law col-md-8 col-lg-9 col-xs-12 topSpace10")
        text = cuad.find_all(text=True)
        
        if 'Attachments' in text:
            for at in str(cuad.find('div',class_="attachments")).split('href="'):
                if '/cld/doc' in at:
                    attach.append('https://sherloc.unodc.org' + at.split('" ')[0])
            
        title = text[1]
    if'/treaties/status/' in url:
        cuad = soup.find_all('div',class_="container")
        text = cuad[3].find_all(text=True)
        title = text[1]
        if 'Attachments' in text:
            print("Attached file not downloaded. " + title)
    elif '/strategies/' in url:
        cuad = soup.find('div',class_="col-md-9 col-lg-9 col-xs-12 topSpace20")
        text = cuad.find_all(text=True)
        
        if 'Attachments' in text:
            for at in str(cuad.find('div',class_="topSpace10")).split('href="'):
                if '/cld/doc' in at:
                    attach.append('https://sherloc.unodc.org' + at.split('" ')[0])
        title=text[0]
    
    while not title[-1].isalpha() and not title[-1].isdigit() and title[-1] != ')':
            title = title[:-1]
    
    title,text = remplace_title_text(title,text)
    return (title, text, attach)

##########################################################################################

def down_attach(url, path, title):
    path = path + '/' + title + '/'
    not_dw = ''
    
    if '.html' in url:
        con=0
        while con == 0:
            try:
                res = requests.get(url)
                con = 1
            except:
                con = 0
                sleep(1)
        html_page = res.content
        soup = BeautifulSoup(html_page, 'lxml')
        cuad = soup.find_all('div',class_="link")
        if len(cuad)==0:
            not_dw = 'Error 404: ' + url
        elif len(cuad) > 1:
            for link in cuad:
                not_dw = not_dw + ',' + 'https://sherloc.unodc.org' + str(link).split('href="')[1].split('" ')[0]
        else:
            attach = 'https://sherloc.unodc.org' + str(cuad[0]).split('href="')[1].split('" ')[0]
    elif '.pdf' in url:
        attach = url
    else:
        not_dw = url

    if not_dw == '':
        title = attach.split('/')[-1].split('.pdf')[0]
        path = path + title
        
        try:
            r = requests.get(attach, stream=True)
            with open(path+'.pdf', 'wb') as f:
                f.write(r.content)
        except:
            not_dw = url
    return not_dw

##########################################################################################

def generate_latex_text(title,text):
    title = title.replace("_", "\_")
    title = title.replace("&","\&")
    latex_text=['\\documentclass{article}',
                '\\usepackage[utf8]{inputenc}',
                '\\title{'+ title +'}',
                '\\author{}',
                '\\date{}',
                '\\begin{document}',
                '\\maketitle']
    for line in text:
        latex_text.append(line)
    end_text='\\end{document}'
    latex_text.append(end_text)
    return latex_text

##########################################################################################

def latex_text_file(latex_text,path,title,url):
    or_title=title
    or_path = path
    title = title.replace("  "," ")
    
    
    if len(title)<100:
        path = or_path + '/' + title
        if not os.path.exists(path):
            os.makedirs(path)
    else:
        sp_tit=title.split(' ')
        vl = sp_tit[6]
        key=0
        for t in sp_tit:
            if t.isdigit()and key==0:
                vl=t
                key=1
                
        title= title[0:title.index(vl)+len(vl)]
        path = or_path+ '/' + title
        if not os.path.exists(path):
            os.makedirs(path)
    
    add = ' 1'
    i = 2
    while os.path.exists(path+'/wPDF '+title+add+'.pdf'):
        add= ' '+str(i)
        i = i+1
    
    title = title+add
    
    for line in latex_text:
        os.system('(echo "' + line + '" >> "' + title + '.tex")')

    os.system('(pdflatex "'+title+'.tex" >/dev/null 2>&1)')
    #-interaction=nonstopmode

    i=0
    OK = 0
    
    Black_List=[]
    
    while OK == 0 and not(os.path.exists(title+'.aux') and os.path.exists(title+'.log') and os.path.exists(title+'.pdf')):
        sleep(1)
        i=i+1
        if i==10:
            print('Error loading: '+or_title+' '*40 +'\n'+ url +'\n')
            OK = 1
    
    
    if OK == 0:
        msg = title + ' OK'
        os.system('rm "'+title+'.aux"')
        os.system('rm "'+title+'.log"')
        os.system('mv "'+title+'.pdf" "'+path+'/wPDF '+title+'.pdf"')
        os.system('rm "'+title+'.tex"')
    elif OK == 1:
        Black_List.append(url)
        msg = title + ' Fail'
        os.system('rm "'+title+'.aux"')
        os.system('rm "'+title+'.log"')
        os.system('mv "'+title+'.tex" "'+title+'.error.tex"')
    
    return Black_List, msg

##########################################################################################
    
def create_fold(ctr,path):
    
    tipes=['Legislation','Strategies','Treaties']
    init_path = path
    Black_List = []
    attachs = []
    not_dw = []
    msgs=[]
    print(ctr+ ': ')
    for ops,tipe in enumerate(tipes):
        op = ops + 1
        urls = extract_urls(ctr,op)
        top= len(urls)
        path = init_path + '/' + tipe
        for val,url in enumerate(urls):
            title,text ,at = load_text(url)
            latex_text=generate_latex_text(title,text)
            bl,msg = latex_text_file(latex_text,path,title,url)
            Black_List = Black_List + bl
            if at != []:
                for a in at:
                    if a not in attachs:
                        nd = down_attach(a, path, title)
                        if nd != '':
                            not_dw.append(nd)
                        attachs.append(a)
                
            msgs.append(msg)
            if len(msgs) > 1:
                print(' '*len(ctr) +' ' * len(msgs[-2])+' ' * 34,end='\r')
            print(' '*len(ctr) + '-' + tipe+ ' ' +str(int(val*100/top))+'% | '+ msgs[-1],end='\r')
        print(' '*len(ctr) + '-' + tipe+ ' 100%' + ' ' * (len(msgs[-1])+4))
    return Black_List, attachs, not_dw

##########################################################################################

def download_lat_files(path,lat_ctrs):
    if not os.path.exists(path):
            os.makedirs(path)
    loaded_ctr=os.listdir(path)
    Black_list=[]
    attach = []
    not_down = []
    for lc in lat_ctrs:
        if lc+'_inc' in loaded_ctr:
            os.system('rm -rf "'+path+ lc + '_inc'+'"')
        if not lc in loaded_ctr:
            print('#########################################')
            inc_ctr = lc + '_inc'
            bl, at, nd = create_fold(lc,path+inc_ctr)
            Black_list= Black_list + bl
            attach = attach + at
            not_down = not_down + nd
            if bl == []:
                os.system('mv "'+path+inc_ctr+'" "'+path+lc+'"')
            else:
                os.system('rm -rf "'+path+inc_ctr+'"')
            print('#########################################\n\n')
    return Black_list, attach, not_down

##########################################################################################

def titles(cuad_str):
    h4=cuad_str.split('<h4 class="law-title ng-binding">')
    vals={}
    for i in h4:
        if '<span class="ng-binding ng-scope"' in i:
            title= i[i.index('</span>')+len('</span>'):i.index('</h4>')]
            title=title[0:-33]
            vals[title]=[]
            sub_titles = i.split('<h5 class="leg-article ng-scope"')
            
            for j in sub_titles:
                elem={}
                if '<span class="ng-binding ng-scope"' in j:
                    sub_sub=j.split('<span class="ng-binding ng-scope"')
                    st=''
                    for k in sub_sub:
                        if '</i></span>' in k:
                            st= st+k[k.index('</i></span>')+len('</i></span>'):k.index('</span><!')]
                    st = st.replace(" \xa0","").replace("&amp;", "&")
                    link=j[j.index('href="')+len('href="'):j.index('"></a>')]
                    elem[st[1:]]= 'https://sherloc.unodc.org' + link
                    
                    vals[title].append(elem)
    return vals
```
